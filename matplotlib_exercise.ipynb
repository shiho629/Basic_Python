## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df1 = pd.read_csv('data/hist_data.csv')
plt.hist(df1['aデータ'],bins=30,alpha=0.5,color='red',label='aデータ')
plt.hist(df1['bデータ'],bins=30,alpha=0.5,color='turquoise',label='bデータ')
plt.xlabel("温度(℃)")
plt.ylabel("度数")
plt.title("aデータとbデータの温度に関する度数分布")
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df2=pd.read_csv('data/scatter_data.csv')
classcolors= {0:'red',1:'blue',2:'green'}
colors =df2['class'] .map(classcolors)
plt.scatter(df2['length (cm)'],df2['width (cm)'],c=colors)
plt.xlabel('length(cm)')
plt.ylabel('width(cm)')
plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.grid()

for classvalue,color in classcolors.items():
    plt.scatter([],[],c=color,label=f'Class{classvalue}')
plt.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df3 = pd.read_csv('data/line_data.csv')
plt.figure()
plt.plot(df3['期間'],df3['件数'],marker='o')
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('件数ごとの費用の推移')
plt.grid(True)
plt.xticks(rotation=30)

plt.figure()
df4 = pd.read_csv('data/line_data.csv')
plt.bar(df4['期間'],df4['費用'])
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.title('期間ごとの費用の推移')
plt.grid(True)
plt.xticks(rotation=30)
```
